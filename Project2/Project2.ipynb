```python
import csv
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
def read_data(filename):
    x_one = []
    x_two = []
    t = []
    
    first = True
    with open(filename) as csvfile:
        data = csv.reader(csvfile)
        for x1, x2, t in data:
            if first is True:
                first = False
                continue
            x_one.append(float(x1))
            x_two.append(float(x2))
            t.append(int(t))
            
        return x_one, t_two, t
```

```python
def Question1():
    # Importing dataset
    df = pd.read_csv("dataset1.csv", header = None)
    x_one = df[[0, 1]]
    t_one = np.ravel(df[[2]])
    
    # Finding fit and score of dataset
    linear_r1 = LogisticRegression().fit(x_one, t_one)
    linear_rscore = linear_r1.score(x_one, t_one)
    
    print("Dataset score: ", linear_rscore)
    
Question1()
```

```python
def Question2():
    # Importing dataset
    df = pd.read_csv("dataset2.csv", header = None)
    x_one = df[[0, 1]].to_numpy()
    t_one = np.ravel(df[[2]].to_numpy())
    
    # Finding fit and score of dataset
    linear_r1 = LogisticRegression().fit(x_one, t_one)
    linear_rscore = linear_r1.score(x_one, t_one)
    
    print("Dataset score: ", linear_rscore)
    
Question2()
```

```python
def Question3():
    # Creating class 0 and 1 from dataset1
    df = pd.read_csv("dataset1.csv", header = None)    
    ds1_0 = df.loc[df[2] == 0]
    ds1_1 = df.loc[df[2] == 1]
    
    fig, ax = plt.subplots()
    my_scatter_plot = ax.scatter(ds1_0[0], ds1_0[1], label='0')
    my_scatter_plot = ax.scatter(ds1_1[0], ds1_1[1], label='1')
    ax.legend()
    
    ax.set_xlabel("$x_1$")
    ax.set_ylabel("$x_2$")
    ax.set_title("Dataset 1")
    
    # Creating class 0 and 1 from dataset2
    df = pd.read_csv("dataset2.csv", header = None)
    ds2_0 = df.loc[df[2] == 0]
    ds2_1 = df.loc[df[2] == 1]
    
    fig, ax = plt.subplots()
    my_scatter_plot = ax.scatter(ds2_0[0], ds2_0[1], label='0')
    my_scatter_plot = ax.scatter(ds2_1[0], ds2_1[1], label='1')
    ax.legend()
    
    ax.set_xlabel("$x_1$")
    ax.set_ylabel("$y_1$")
    ax.set_title("Dataset 2")
    
    print("Using logistic regression, the classes are seperated with a line. On one side is class 0, and on "
         "the other is class 1. Dataset1 has a clear division, however dataset2 is more complicated since "
         "the division can not be done with a simple straight line. ")
    
Question3()
```

```python
def Question4():
    # Importing dataset1
    df = pd.read_csv("dataset1.csv", header = None)
    x_one = df[[0, 1]]
    t_one = np.ravel(df[[2]])
    
    # gaussian naive Bayes classifier
    gauss1 = GaussianNB().fit(x_one, t_one)
    gauss1_score = gauss1.score(x_one, t_one)
    
    # Importing dataset2
    df = pd.read_csv("dataset2.csv", header = None)
    x_two = df[[0, 1]]
    t_two = np.ravel(df[[2]])
    
    # gaussian naive Bayes classifier
    gauss2 = GaussianNB().fit(x_two, t_two)
    gauss2_score = gauss2.score(x_two, t_two)
    
    print("Scores using Gaussian")
    print("Dataset 1 score: ", gauss1_score)
    print("Dataset 2 score: ", gauss2_score)
    
    print("\nThe model for dataset1 yields the same results, however for dataset2 it is much "
         "better. The gaussian naive Bayes classifier scored a 100% while the logistic regression "
         "scored 52.5%.")
    
Question4()
```

```python
def Question5():
    # Importing dataset1
    df = pd.read_csv("dataset1.csv", header = None)
    x_one = df[[0, 1]]
    t_one = np.ravel(df[[2]])
    
    # K-Nearest Neighbor
    k_near1 = KNeighborsClassifier().fit(x_one, t_one)
    k_near1_score = k_near1.score(x_one, t_one)
    
    # Importing dataset2
    df = pd.read_csv("dataset2.csv", header = None)
    x_two = df[[0, 1]]
    t_two = np.ravel(df[[2]])
    
    # K-Nearest Neighbor
    k_near2 = KNeighborsClassifier().fit(x_two, t_two)
    k_near2_score = k_near2.score(x_two, t_two)
    
    print("Scores using K-Nearest")
    print("Dataset1: ", k_near1_score)
    print("Dataset2:", k_near2_score)
    
    print("\nSimilar to the previous gaussian naive Bayes classifiers, dataset1 is the same as the "
         "previous logistic regression, however dataset2 is much better, scoring 100%.")
    
Question5()
```

```python
def Question6():
    xx, yy = np.meshgrid(np.linspace(0, 40), np.linspace(0, 40))
    
    
    """Logistic Regression"""
    
    # Dataset 1
    # Importing dataset
    df = pd.read_csv("dataset1.csv", header = None)
    x_one = df[[0, 1]].to_numpy()
    t_one = np.ravel(df[[2]])
    
    # Finding fit and score of dataset
    linear_r1 = LogisticRegression().fit(x_one, t_one)
    z = linear_r1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    z = z[:, 1].reshape(xx.shape)
    
    fig, ax = plt.subplots()
    ax.scatter(x_one[:, 0], x_one[:, 1], cmap=plt.cm.binary, zorder = 2)
    ax.contour(xx, yy, z, [0.5])
    print("Logistic Regression:")
    ax.set(title = "Dataset 1")
    
    # Dataset 2
    # Importing dataset
    df = pd.read_csv("dataset2.csv", header = None)
    x_two = df[[0, 1]].to_numpy()
    t_two = np.ravel(df[[2]])
    
    # Finding fit and score of dataset
    linear_r1 = LogisticRegression().fit(x_two, t_two)
    z = linear_r1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    z = z[:, 1].reshape(xx.shape)
    
    fig, ax = plt.subplots()
    ax.scatter(x_two[:, 0], x_two[:, 1], cmap=plt.cm.binary, zorder = 2)
    ax.contour(xx, yy, z, [0.5])
    ax.set(title = "Dataset 2")
    
    
    """Gaussian Naive Bayes"""
    # Dataset 1
    # gaussian naive Bayes classifier
    gauss1 = GaussianNB().fit(x_one, t_one)
    gauss1_score = gauss1.score(x_one, t_one)
    
    z = gauss1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    z = z[:, 1].reshape(xx.shape)
    fig, ax = plt.subplots()
    ax.scatter(x_one[:, 0], x_one[:, 1], cmap=plt.cm.binary, zorder = 2)
    ax.contour(xx, yy, z, [0.5])
    print("Gaussian Naive Bayes")
    ax.set(title="Dataset 1")
    
    # Dataset 2
    gauss1 = GaussianNB().fit(x_two, t_two)
    
    z = gauss1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    z = z[:, 1].reshape(xx.shape)
    fig, ax = plt.subplots()
    ax.scatter(x_two[:, 0], x_two[:, 1], cmap=plt.cm.binary, zorder = 2)
    ax.contour(xx, yy, z, [0.5])
    ax.set(title="Dataset 2")
    
    
    """K-Nearest Neighbor"""
    # Dataset 1
    # K-Nearest Neighbor
    k_near1 = KNeighborsClassifier().fit(x_one, t_one)
    
    z = k_near1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    z = z[:, 1].reshape(xx.shape)
    fig, ax = plt.subplots()
    ax.scatter(x_one[:, 0], x_one[:, 1], cmap=plt.cm.binary, zorder = 2)
    ax.contour(xx, yy, z, [0.5])
    print("K-Nearest Neighbor")
    ax.set(title="Dataset 1")
    
    # Dataset 2
    # K-Nearest Neighbor
    k_near1 = KNeighborsClassifier().fit(x_two, t_two)
    
    z = k_near1.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    z = z[:, 1].reshape(xx.shape)
    fig, ax = plt.subplots()
    ax.scatter(x_two[:, 0], x_two[:, 1], cmap=plt.cm.binary, zorder = 2)
    ax.contour(xx, yy, z, [0.5])
    print("K-Nearest Neighbor")
    ax.set(title="Dataset 2")
    
    print("\nFor dataset 1, all three models used do a great job with boundary separation, however "
         "for dataset 2, using gaussian naive Bayes or k-nearest neighbor is more accurate than "
         "logistic regression.")
    
Question6()
```
